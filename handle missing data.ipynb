```python
import pandas as pd

india_weather = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
us_weather
```

```python
df = pd.concat([india_weather, us_weather])
df
```

```python
df = pd.concat([india_weather, us_weather], ignore_index=True)
df
```

```python
df = pd.concat([india_weather, us_weather], keys=["india", "us"])
df
```

```python
temperature_df = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
}, index=[0,1,2])
temperature_df
```

```python
windspeed_df = pd.DataFrame({
    "city": ["delhi","mumbai"],
    "windspeed": [7,12],
}, index=[1,0])
windspeed_df
```

```python
#Default axis is 0 that denotes rows if you want to concat column wise use axis=1

df = pd.concat([temperature_df,windspeed_df],axis=1)
df
```

```python
s = pd.Series(["Humid","Dry","Rain"], name="event")
s
```

```python
df = pd.concat([temperature_df,s],axis=1)
df
```

```python

df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35, 38],
})
df1
```

```python

df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})
df2
```

```python
# how denotes type of join it refers to. It can be left,right,inner whereas indicator denotes 
# the _merge column that tell src of data

df3=pd.merge(df1,df2,on="city",how="outer",indicator=True)
df3
```
